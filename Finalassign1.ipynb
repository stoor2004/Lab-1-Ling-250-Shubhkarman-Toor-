```python
from nltk.corpus import gutenberg
from nltk.tokenize import sent_tokenize

# Load the Gutenberg corpus
corpus = gutenberg.raw()

# Tokenize the sentences
sentences = sent_tokenize(corpus)
```

```python
# Find the longest sentence
longest_sentence = max(sentences, key=lambda sentence: len(nltk.word_tokenize(sentence)))

# Get the number of words in the longest sentence
num_words = len(nltk.word_tokenize(longest_sentence))

# Print the longest sentence and the number of words
print("Longest Sentence:")
print(longest_sentence)
print("\nNumber of Words:", num_words)
```
